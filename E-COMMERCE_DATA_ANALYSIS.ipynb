## Groupby Analytics


This stage aggregates data using groupby operations to understand trends across categories, age groups, genders, and products.


```python
df['category_revenue']=df.groupby('Category')['revenue'].transform('sum')
```

```python
summary = df.groupby('Category')['revenue'].sum().reset_index()

summary['revenue_share'] = (
    summary['revenue'] / summary['revenue'].sum()
) * 100

summary.sort_values(by='revenue_share', ascending=False)
```

```python
summary = df.groupby('Customer Gender')['revenue'].sum().reset_index()

summary['revenue_share'] = (
    summary['revenue'] / summary['revenue'].sum()
) * 100

summary.sort_values(by='revenue_share', ascending=False)
```

```python
summary = df.groupby('Customer Age Group')['revenue'].sum().reset_index()

summary['revenue_share'] = (
    summary['revenue'] / summary['revenue'].sum()
) * 100

summary.sort_values(by='revenue_share', ascending=False)
```

```python
summary = df.groupby('Product Name')['revenue'].sum().reset_index()

summary['revenue_share'] = (
    summary['revenue'] / summary['revenue'].sum()
) * 100

summary.sort_values(by=['revenue_share'], ascending=False)
```

## Multiple Visualizations
The following plots visually explain the analytical insights derived from the dataset such as category revenue, demand patterns, and product performance.


```python
import matplotlib.pyplot as plt

spmethod = df['Shipping Method'].value_counts()

plt.figure(figsize=(6,5))
plt.bar(spmethod.index, spmethod.values)

plt.title('Shipping Method Distribution')
plt.xlabel('Shipping Method')
plt.ylabel('Number of Orders')

plt.show()
```

```python
x=topsales['Product Name']
y=topsales['revenue_share']
plt.figure(figsize=(8,7))
plt.bar(x, y)

plt.title('top 10 products by thier revenue')
plt.xlabel('products')
plt.ylabel(' Revenue')

plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
plt.style.use('ggplot')
summary = df.groupby('Customer Age Group')['revenue'].sum()
x = summary.index
y = summary.values
plt.plot(x,y,marker='o',color='g',label='Developers')
plt.legend()
plt.title('myplot')
plt.xlabel('AGE')
plt.ylabel('REVENUE')
plt.grid(True)
plt.show()
```

```python
plt.style.use('ggplot')
summary = df.groupby('Customer Age Group')['revenue'].sum()
x = summary.index
y = summary.values
plt.plot(x,y,marker='o',color='g',label='Developers')
plt.legend()
plt.title('myplot')
plt.xlabel('AGE')
plt.ylabel('REVENUE')
plt.grid(True)
plt.show()
```

```python
category_revenue = df.groupby('Category')['revenue'].sum()
plt.figure(figsize=(5,7))
plt.bar(category_revenue.index, category_revenue.values)

plt.title('Category-wise Total Revenue')
plt.xlabel('Category')
plt.ylabel('Total Revenue')

plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
popularity = df['Popularity Index']
stocklevel = df['Stock Level']

ratio = popularity / stocklevel

plt.scatter(popularity, stocklevel,
            c=ratio,
            cmap='summer',
            edgecolor='black',
            linewidth=1,
            alpha=0.75)

clrbar = plt.colorbar()
clrbar.set_label('Popularity/Stock Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Popularity vs Stock Level')
plt.xlabel('Popularity Index')
plt.ylabel('Stock Level')

plt.tight_layout()
plt.show()
```
